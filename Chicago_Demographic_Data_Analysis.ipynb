## III. Analytical tasks


### Task 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%sql SELECT COUNT(CASE_NUMBER) FROM CHICAGO_CRIME_DATA
```

### Task 2

##### List community area names and numbers with per capita income less than 11000.


```python
%%sql 
SELECT COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER, PER_CAPITA_INCOME 
FROM CENSUS_DATA 
WHERE PER_CAPITA_INCOME < 11000
```

### Task 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
%%sql 
SELECT PRIMARY_TYPE, DESCRIPTION 
FROM CHICAGO_CRIME_DATA 
WHERE DESCRIPTION LIKE "%MINOR%"
```

### Task 4

##### List all kidnapping crimes involving a child


```python
%%sql 
SELECT PRIMARY_TYPE, DESCRIPTION 
FROM CHICAGO_CRIME_DATA 
WHERE PRIMARY_TYPE = "KIDNAPPING" AND DESCRIPTION LIKE "%CHILD%"
```

### Task 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
%%sql 
SELECT DISTINCT PRIMARY_TYPE, DESCRIPTION, LOCATION_DESCRIPTION 
FROM CHICAGO_CRIME_DATA 
WHERE LOCATION_DESCRIPTION LIKE "%SCHOOL%"
```

### Task 6

##### List the type of schools along with the average safety score for each type.


```python
%%sql 
SELECT "Elementary, Middle, or High School" AS school_type, AVG(SAFETY_SCORE) AS avg_safety_score 
FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY school_type
```

### Task 7

##### List 5 community areas with highest % of households below poverty line


```python
%%sql 
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY 
FROM CENSUS_DATA 
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY 
DESC LIMIT 5
```

### Task 8

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql
SELECT COMMUNITY_AREA_NAME
FROM CENSUS_DATA 
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA)
```

<u>Reference table used for tasks 9 and 10</u>

```python
%%sql
SELECT CS.COMMUNITY_AREA_NUMBER, CS.COMMUNITY_AREA_NAME, COUNT(CR.CASE_NUMBER) AS number_of_cases
FROM CENSUS_DATA CS
JOIN CHICAGO_CRIME_DATA CR ON CS.COMMUNITY_AREA_NUMBER = CR.COMMUNITY_AREA_NUMBER
GROUP BY CS.COMMUNITY_AREA_NUMBER
```

### Task 9

##### Which community area is most crime prone? Display the coumminty area number only.


```python
%%sql
WITH area_vs_crime AS (
  SELECT CS.COMMUNITY_AREA_NUMBER, COUNT(CR.CASE_NUMBER) AS number_of_cases
  FROM CENSUS_DATA CS
  JOIN CHICAGO_CRIME_DATA CR ON CS.COMMUNITY_AREA_NUMBER = CR.COMMUNITY_AREA_NUMBER
  GROUP BY CS.COMMUNITY_AREA_NUMBER
)
SELECT community_area_number
FROM area_vs_crime
WHERE number_of_cases = (SELECT MAX(number_of_cases) FROM area_vs_crime);
```

### Task 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
WITH area_vs_crime AS (
  SELECT CS.COMMUNITY_AREA_NUMBER, CS.COMMUNITY_AREA_NAME, COUNT(CR.CASE_NUMBER) AS number_of_cases
  FROM CENSUS_DATA CS
  JOIN CHICAGO_CRIME_DATA CR ON CS.COMMUNITY_AREA_NUMBER = CR.COMMUNITY_AREA_NUMBER
  GROUP BY CS.COMMUNITY_AREA_NUMBER
)
SELECT COMMUNITY_AREA_NAME
FROM area_vs_crime
WHERE number_of_cases = (SELECT MAX(number_of_cases) FROM area_vs_crime);
```
