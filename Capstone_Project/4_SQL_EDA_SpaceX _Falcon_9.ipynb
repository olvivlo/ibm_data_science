## Tasks

### Task 1: Display the names of the unique launch sites  in the space mission

```python
%%sql 
SELECT DISTINCT Launch_Site FROM SPACEXTBL 
```


### Task 2: Display 5 records where launch sites begin with the string 'CCA' 


```python
%%sql 
SELECT *
FROM SPACEXTBL 
WHERE Launch_Site LIKE "%CCA%" 
LIMIT 5
```

### Task 3: Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%%sql 
SELECT *
FROM SPACEXTBL 
WHERE Customer =  "NASA (CRS)"
```

```python
%%sql 
SELECT SUM(PAYLOAD_MASS__KG_)
FROM SPACEXTBL 
WHERE Customer =  "NASA (CRS)"
```

### Task 4: Display average payload mass carried by booster version F9 v1.1


```python
%%sql 
SELECT *
FROM SPACEXTBL 
WHERE Booster_Version =  "F9 v1.1"
```

```python
%%sql 
SELECT AVG(PAYLOAD_MASS__KG_)
FROM SPACEXTBL 
WHERE Booster_Version =  "F9 v1.1"
```

### Task 5: List the date when the first succesful landing outcome in ground pad was acheived

```python
%%sql 
SELECT *
FROM SPACEXTBL 
WHERE Landing_Outcome = "Success (ground pad)"
```

```python
%%sql 
SELECT MIN(Date)
FROM SPACEXTBL 
WHERE Landing_Outcome = "Success (ground pad)"
```

### Task 6: List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
%%sql 
SELECT *
FROM SPACEXTBL 
WHERE Landing_Outcome = "Success (drone ship)"
```

```python
%%sql 
SELECT Booster_Version, PAYLOAD_MASS__KG_
FROM SPACEXTBL 
WHERE Landing_Outcome = "Success (drone ship)" 
    AND PAYLOAD_MASS__KG_ > 4000
    AND PAYLOAD_MASS__KG_ < 6000
```

### Task 7: List the total number of successful and failure mission outcomes


```python
%%sql 
SELECT Mission_Outcome
FROM SPACEXTBL 
GROUP BY Mission_Outcome
```

```python
%%sql 
SELECT
    CASE
        WHEN TRIM(Mission_Outcome) LIKE "Success%" THEN "Success"
        WHEN TRIM(Mission_Outcome) LIKE "Failure%" THEN "Failure"
        ELSE 'Other'
    END AS Clean_Mission_Outcome,
    COUNT(*)
FROM SPACEXTBL 
GROUP BY Clean_Mission_Outcome;
```

### Task 8: List the   names of the booster_versions which have carried the maximum payload mass


```python
%%sql 
SELECT DISTINCT(Booster_Version)
FROM SPACEXTBL 
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL)
```

### Task 9: List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015


```python
%%sql
SELECT *
FROM SPACEXTBL
WHERE SUBSTR(Date,1,4) = '2015'
```

```python
%%sql
SELECT SUBSTR(Date, 6, 2) AS Month
FROM SPACEXTBL
WHERE SUBSTR(Date,1,4) = '2015'
```

```python
%%sql
SELECT
       CASE SUBSTR(Date, 6, 2)
           WHEN '01' THEN 'January'
           WHEN '02' THEN 'February'
           WHEN '03' THEN 'March'
           WHEN '04' THEN 'April'
           WHEN '05' THEN 'May'
           WHEN '06' THEN 'June'
           WHEN '07' THEN 'July'
           WHEN '08' THEN 'August'
           WHEN '09' THEN 'September'
           WHEN '10' THEN 'October'
           WHEN '11' THEN 'November'
           WHEN '12' THEN 'December'
       END AS Month,
    Landing_Outcome,
    Booster_Version, 
    Launch_Site
FROM SPACEXTBL
WHERE SUBSTR(Date,1,4) = '2015' 
```

```python
%%sql
SELECT
       CASE SUBSTR(Date, 6, 2)
           WHEN '01' THEN 'January'
           WHEN '02' THEN 'February'
           WHEN '03' THEN 'March'
           WHEN '04' THEN 'April'
           WHEN '05' THEN 'May'
           WHEN '06' THEN 'June'
           WHEN '07' THEN 'July'
           WHEN '08' THEN 'August'
           WHEN '09' THEN 'September'
           WHEN '10' THEN 'October'
           WHEN '11' THEN 'November'
           WHEN '12' THEN 'December'
       END AS Month,
    Landing_Outcome,
    Booster_Version, 
    Launch_Site
FROM SPACEXTBL
WHERE SUBSTR(Date,1,4) = '2015' 
    AND SPACEXTBL.Landing_Outcome = 'Failure (drone ship)';
```

### Task 10: Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
%%sql
SELECT *
FROM SPACEXTBL
WHERE Date >= '2010-06-04'
    AND Date < '2017-03-20'
```

```python
%%sql
SELECT Landing_Outcome, COUNT(Landing_Outcome) AS Count
FROM SPACEXTBL
WHERE Date >= '2010-06-04'
    AND Date < '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY Count DESC
```
